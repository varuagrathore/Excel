```python
print(df.columns)
print(new_data.columns)
# Merge the two DataFrames on 'Content ID'
merged_df = df.merge(new_data[['Content ID', 'Content Type', 'Category']], on='Content ID', how='inner')

# Print the merged DataFrame
merged_df


```

```python
merged_df.columns
```

```python
final_merged_df = merged_df.merge(data_reaction_type, on='Reaction_Type', how='inner')
```

```python
final_merged_df
```

```python

# Calculate total scores for each category
category_scores = final_merged_df.groupby('Category')['Score'].sum().reset_index()

# Sort the categories based on total scores in descending order
top_5_categories = category_scores.sort_values(by='Score', ascending=False)

# Print the top 5 performing categories
print(top_5_categories)
```

```python
# Save the cleaned dataset to a CSV file
final_merged_df.to_csv('cleaned_dataset.csv', index=False)
```

```python
top_5_categories.to_csv('top5.csv',index=False)
```
